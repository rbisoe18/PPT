#### Q. 1

```python
class ListNode:
    def __init__(self, x):
        self.val = x
        self.next = None

def detectCycle(head):
    # Check if the linked list is empty or has only one node
    if not head or not head.next:
        return None

    slow = head
    fast = head

    # Move slow and fast pointers until they meet
    while fast and fast.next:
        slow = slow.next
        fast = fast.next.next

        # If the two pointers meet, there is a cycle
        if slow == fast:
            break

    # If the fast pointer reached the end, there is no cycle
    if fast is None or fast.next is None:
        return None

    # Move one pointer back to the head
    slow = head

    # Move both pointers at the same speed until they meet again
    while slow != fast:
        slow = slow.next
        fast = fast.next

    # Return the node where the cycle begins
    return slow

# Example 1
head1 = ListNode(3)
head1.next = ListNode(2)
head1.next.next = ListNode(0)
head1.next.next.next = ListNode(-4)
head1.next.next.next.next = head1.next

cycle_start1 = detectCycle(head1)
if cycle_start1:
    print("Output: tail connects to node index", cycle_start1.val)
else:
    print("Output: no cycle")
    
print("\n---------------------------------")
# Example 2
head2 = ListNode(1)
head2.next = ListNode(2)
head2.next.next = head2

cycle_start2 = detectCycle(head2)
if cycle_start2:
    print("Output: tail connects to node index", cycle_start2.val)
else:
    print("Output: no cycle")
print("\n---------------------------------")
# Example 3
head3 = ListNode(1)

cycle_start3 = detectCycle(head3)
if cycle_start3:
    print("Output: tail connects to node index", cycle_start3.val)
else:
    print("Output: no cycle")
```

```python
class ListNode:
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next

def pairSum(head):
    # Find the middle of the linked list
    slow = head
    fast = head
    maxVal = 0
    while fast and fast.next:
        slow = slow.next
        fast = fast.next.next

    # Reverse the second half of the linked list
    prev = None
    while slow:
        next_node = slow.next
        slow.next = prev
        prev = slow
        slow = next_node

    # Compare the values of the first half with the reversed second half
    while prev:
        maxVal = max(maxVal, head.val + prev.val)
        prev = prev.next
        head = head.next
    return maxVal
    return True
# Example 1
head1 = ListNode(5)
head1.next = ListNode(4)
head1.next.next = ListNode(2)
head1.next.next.next = ListNode(1)

print(pairSum(head1))

print("----")
# Example 2
head2 = ListNode(4)
head2.next = ListNode(2)
head2.next.next = ListNode(2)
head2.next.next.next = ListNode(3)

print(pairSum(head2))

print("----")

# Example 3
head3 = ListNode(1)
head3.next = ListNode(100000)

print(pairSum(head3))
```

```python
class ListNode:
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next


def removeNthFromEnd(head, n):
    dummy = ListNode(0)  # Create a dummy node to handle edge cases
    dummy.next = head
    slow = dummy
    fast = dummy

    # Move the fast pointer ahead by n positions
    for _ in range(n):
        fast = fast.next

    # Move both pointers until fast reaches the end
    while fast.next is not None:
        slow = slow.next
        fast = fast.next

    # Skip the nth node by updating the next reference of the slow pointer
    slow.next = slow.next.next

    return dummy.next
# Example 1
head1 = ListNode(1)
head1.next = ListNode(2)
head1.next.next = ListNode(3)
head1.next.next.next = ListNode(4)
head1.next.next.next.next = ListNode(5)
n1 = 2

result1 = removeNthFromEnd(head1, n1)

output1 = []
current1 = result1
while current1 is not None:
    output1.append(current1.val)
    current1 = current1.next

print(output1)  # Output: [1, 2, 3, 5]

print("----------")

# Example 2
head2 = ListNode(1)
n2 = 1

result2 = removeNthFromEnd(head2, n2)

output2 = []
current2 = result2
while current2 is not None:
    output2.append(current2.val)
    current2 = current2.next

print(output2)  # Output: []

print("----------")

# Example 3
head3 = ListNode(1)
head3.next = ListNode(2)
n3 = 1

result3 = removeNthFromEnd(head3, n3)

output3 = []
current3 = result3
while current3 is not None:
    output3.append(current3.val)
    current3 = current3.next

print(output3)  # Output: [1]
```
